```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# import dataset
ds=pd.read_csv("new_cleaned_loan_data.csv")
ds
```

```python
ds.info()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(ds['loan_amount'], kde=True, color='skyblue')
plt.title("Distribution of Loan Amount")
plt.xlabel("Loan Amount")
plt.ylabel("Frequency")
plt.show()
```

Interpretation: Most of people are taking small loans

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='loan_status', data=ds, palette='Set2')
plt.title("Loan Status Distribution")
plt.xlabel("Loan Status")
plt.ylabel("Count")
plt.show()
```

Most of the loans are approved 

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='income_annum', y='loan_amount', hue='loan_status', data=ds)
plt.title("Income vs Loan Amount")
plt.xlabel("Annual Income")
plt.ylabel("Loan Amount")
plt.show()
```

higher income leads to larger loans

```python
plt.figure(figsize=(7, 4))
sns.boxplot(x=ds['loan_amount'], color='orange')
plt.title("Boxplot of Loan Amount")
plt.xlabel("Loan Amount")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(ds['cibil_score'], kde=True, color='green')
plt.title("Distribution of CIBIL Score")
plt.xlabel("CIBIL Score")
plt.ylabel("Frequency")
plt.show()
```

most people have average scores between 500-700 ranges 

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='loan_status', y='residential_assets_value', data=ds)
plt.title("Residential Assets by Loan Status")
plt.xlabel("Loan Status")
plt.ylabel("Residential Assets Value")
plt.show()
```

this shows if higher assest values lead to higher chances of loan approvaed

```python
plt.figure(figsize=(10, 6))
sns.heatmap(ds.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

A value closer to 1 or -1 means strong correlation
